```python
from matplotlib import pyplot as plt
import numpy as np

plt.rc('font', family='NanumBarunGothic')
#폰트설정

year = ['2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020']
index = np.arange(len(year))
values1 = [68819, 64823, 64193, 57643, 53776, 52005, 48261, 44746]
#결혼
values2 = [20126, 19477, 18176, 17777, 17083, 16940, 16972, 16282]
#이혼

plt.title('연도별 혼인 이혼 건수')
#제목

bar_width = 0.2
plt.bar(index, values1, bar_width, label='혼인 건수')
plt.bar(index+bar_width, values2, bar_width, label='이혼 건수')
#2줄 막대그래프로 만듦

plt.xticks(index, year)
plt.yticks(values1)

plt.xlabel('년도(2013-2020)', fontsize=10)
plt.ylabel('혼인/이혼 수(건)', fontsize=10)
#라벨링

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
#범례

plt.show()
```
